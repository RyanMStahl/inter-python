```python
class Thing:
    def __init__(self, a=0, b=0):
        self.a = a
        self.b = b

thing = Thing(4,5)
thing.a, thing.b
```

```python
thing.a = 7
thing.a, thing.b
```

```python
thing
```

```python
things = []
for i in range(0,100):
    things.append(Thing(i,i*2))

things[50].a,things[50].b
```

```python
import pickle

with open("things.pkl","wb") as f:
    pickle.dump(things,f)
    
```

```python
with open("things.pkl","rb") as f:
    things2 = pickle.load(f)

things2[0],things2[50].a
```
